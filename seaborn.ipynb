```python
df = sb.load_dataset('flights')
df
```

### _RELPLOT_

```python
sb.relplot(x = 'year', y = 'passengers', data = df, 
           hue = 'month',
          size = 'month')
plt.show()
```

```python
a = sb.load_dataset('flights')
a = a[a['month'] == 'January']
a = a[a['year'] < 1955]
sb.relplot(x = 'year', y = 'passengers', data = a,
           hue = 'year',
          size = 'year',
#           style = 'year'
          )

plt.yticks(rotation=90)
plt.ylabel('penumpang')

plt.show()
```

### _BARPLOT_

```python
sb.barplot(
    x = 'year', y = 'passengers', data = df,
    hue='month')
plt.legend(loc=2, prop={'size':7.5})
plt.show()
```

### _CATPLOT_

```python
sb.catplot(
    x = 'year', y = 'passengers', data = df,
    hue='month')

plt.show()
```

### _HEATMAP_

```python
sb.heatmap(df.corr(), cmap='hot', annot=True)
```

```python
sb.heatmap(df.pivot('month', 'year', 'passengers'), cmap='hot_r',
           annot=True, fmt='d')
# fmt = f for float, d for integer, 1f for float(1 digit after point)
plt.show()
```

### _VIOLINPLOT_

```python
sb.violinplot(
    x='year', y='passengers', data = df)
plt.show()
```
